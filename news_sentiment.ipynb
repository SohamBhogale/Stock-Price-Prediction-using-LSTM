```python
import csv
from flair.models import TextClassifier
from flair.data import Sentence
import re
import pandas as pd
```

```python
classifier = TextClassifier.load('en-sentiment')
reg = re.compile("[\(\)\]\[]")
```

```python
amazon_sentiments = {}
classifier = TextClassifier.load('en-sentiment')
with open("amazon.csv") as f:
    data = csv.reader(f)
    for each in data:
        date = each[0][:10]
        sentence = Sentence(each[1])
        classifier.predict(sentence)
        labs = sentence.labels
        splitted = str(labs).split()
        sentiment = 0.0
        if 'NEGATIVE' in splitted[0]:
            sentiment = -float(reg.sub("", splitted[1]))
        else:
            sentiment = float(reg.sub("", splitted[1]))
        if date in amazon_sentiments:
            amazon_sentiments[date] = (amazon_sentiments[date][0]+1, (amazon_sentiments[date][1] + sentiment))
        else:
            amazon_sentiments[date] = (1, sentiment)
    f.close()
```

```python
google_sentiments = {}
with open("google.csv") as f:
    data = csv.reader(f)
    for each in data:
        if len(each) > 0 and len(each[0]) > 0:
            date = each[0][:10]
            sentence = Sentence(each[1])
            classifier.predict(sentence)
            labs = sentence.labels
            splitted = str(labs).split()
            sentiment = 0.0
            if 'NEGATIVE' in splitted[0]:
                sentiment = -float(reg.sub("", splitted[1]))
            else:
                sentiment = float(reg.sub("", splitted[1]))
            if date in google_sentiments:
                google_sentiments[date] = (google_sentiments[date][0]+1, (google_sentiments[date][1] + sentiment))
            else:
                google_sentiments[date] = (1, sentiment)
    f.close()
```

```python
fb_sentiments = {}
with open("facebook.csv") as f:
    data = csv.reader(f)
    for each in data:
        print(each)
        if len(each) == 2:
            date = each[0][:10]
            sentence = Sentence(each[1])
            classifier.predict(sentence)
            labs = sentence.labels
            splitted = str(labs).split()
            sentiment = 0.0
            if 'NEGATIVE' in splitted[0]:
                sentiment = -float(reg.sub("", splitted[1]))
            else:
                sentiment = float(reg.sub("", splitted[1]))
            if date in fb_sentiments:
                fb_sentiments[date] = (fb_sentiments[date][0]+1, (fb_sentiments[date][1] + sentiment))
            else:
                fb_sentiments[date] = (1, sentiment)
    f.close()
```

```python
apple_sentiments = {}
with open("apple.csv") as f:
    data = csv.reader(f)
    for each in data:
        if len(each) == 2:
            date = each[0][:10]
            sentence = Sentence(each[1])
            classifier.predict(sentence)
            labs = sentence.labels
            splitted = str(labs).split()
            sentiment = 0.0
            if 'NEGATIVE' in splitted[0]:
                sentiment = -float(reg.sub("", splitted[1]))
            else:
                sentiment = float(reg.sub("", splitted[1]))
            if date in apple_sentiments:
                apple_sentiments[date] = (apple_sentiments[date][0]+1, (apple_sentiments[date][1] + sentiment))
            else:
                apple_sentiments[date] = (1, sentiment)
    f.close()
```

```python
for each in amazon_sentiments.items():
    amazon_sentiments[each[0]] = each[1][1]/each[1][0]
for each in google_sentiments.items():
    google_sentiments[each[0]] = each[1][1]/each[1][0]
for each in apple_sentiments.items():
    apple_sentiments[each[0]] = each[1][1]/each[1][0]
for each in fb_sentiments.items():
    fb_sentiments[each[0]] = each[1][1]/each[1][0]
```

```python
print(len(amazon_sentiments), len(google_sentiments), len(apple_sentiments), len(fb_sentiments))
```

```python
with open('amazon_data.csv') as f:
    data = csv.reader(f)
    dataNew = []
    for i, each in enumerate(data):
        if i == 0:
            dataNew.append(each + [amazon_sentiments.get(each[0],"Sentiment")])
        else:
            dataNew.append(each + [amazon_sentiments.get(each[0],0.0)])
    amazon_data= pd.DataFrame(dataNew, columns =['date', 'open', 'high', 'low', 'close', 'volume', 'Name', 'Sentiment'])
    amazon_data.to_excel('amazon_final.xlsx', header=False, index=False)
    f.close()
```
